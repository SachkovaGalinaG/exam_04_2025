6. state: область или регион.
7. postal_code: почтовый индекс.
8. region: более широкая географическая зона или регион.

# Отчет по возвратам

```python
orders_df.head()
```

```python
orders_df['returned'] = orders_df['returned'].fillna(0)
returns_df = orders_df[orders_df['returned'] == 1]
non_returns_df = orders_df[orders_df['returned'] == 0]
```

```python
total_orders = len(orders_df)
total_returns = len(returns_df)
return_rate = total_returns / total_orders * 100

print(f"Всего заказов: {total_orders}")
print(f"Всего возвратов: {total_returns}")
print(f"Процент возвратов: {return_rate:.2f}%")

# Визуализация
plt.figure(figsize=(8, 5))
sns.countplot(x='returned', data=orders_df)
plt.title('Распределение возвратов')
plt.xlabel('Возврат')
plt.ylabel('Количество')
plt.xticks([0, 1], ['Нет возврата', 'Есть возврат'])
plt.show()
```

```python
# Преобразуем даты в datetime
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])
orders_df['ship_date'] = pd.to_datetime(orders_df['ship_date'])

# Добавим месяц и год заказа
orders_df['order_month'] = orders_df['order_date'].dt.to_period('M')
returns_by_month = orders_df.groupby('order_month')['returned'].mean().reset_index()
returns_by_month['order_month'] = returns_by_month['order_month'].astype(str)

# Визуализация
plt.figure(figsize=(12, 6))
sns.lineplot(x='order_month', y='returned', data=returns_by_month)
plt.title('Динамика процента возвратов по месяцам')
plt.xlabel('Месяц заказа')
plt.ylabel('Доля возвратов')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
returns_by_ship_mode = orders_df.groupby('ship_mode')['returned'].agg(['mean', 'count']).reset_index()
returns_by_ship_mode.columns = ['ship_mode', 'return_rate', 'total_orders']
# Визуализация
plt.figure(figsize=(10, 6))
sns.barplot(x='ship_mode', y='return_rate', data=returns_by_ship_mode)
plt.title('Процент возвратов по способу доставки')
plt.xlabel('Способ доставки')
plt.ylabel('Доля возвратов')
plt.show()
```

```python
# Сравнение средних показателей
metrics = ['sales', 'quantity', 'discount', 'profit']
comparison = orders_df.groupby('returned')[metrics].mean().T

# Визуализация
comparison.plot(kind='bar', figsize=(12, 6))
plt.title('Сравнение средних показателей для возвратов и обычных заказов')
plt.ylabel('Среднее значение')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Рассчитаем время доставки в днях
orders_df['delivery_time'] = (orders_df['ship_date'] - orders_df['order_date']).dt.days

# Сравним время доставки для возвратов и обычных заказов
plt.figure(figsize=(10, 6))
sns.boxplot(x='returned', y='delivery_time', data=orders_df)
plt.title('Время доставки для возвратов и обычных заказов')
plt.xlabel('Возврат')
plt.ylabel('Дней до доставки')
plt.xticks([0, 1], ['Нет возврата', 'Есть возврат'])
plt.show()
```

```python
# Создаем сводную таблицу по способу доставки и размеру скидки
pivot_table = pd.pivot_table(orders_df,
                            values='returned',
                            index='ship_mode',
                            columns=pd.cut(orders_df['discount'], bins=[0, 0.1, 0.3, 0.5, 1]),
                            aggfunc='mean')

# Визуализация
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".2%", cmap='coolwarm')
plt.title('Процент возвратов по способу доставки и размеру скидки')
plt.xlabel('Размер скидки')
plt.ylabel('Способ доставки')
plt.show()
```
