```python
df1=pd.read_csv('sales_of_month.csv')
df1.head()
```

```python
#информация о типе данных каждого столбца
df1.info()
```

```python
#Получение сводной статистики для каждого столбца
df1.describe()
```

```python
df1['returned']=df1['returned'].str.replace(r"Yes","1")
```

```python
#замена значений NaN на 0
df1 = df1.fillna(0)
```

```python
df1
```

Анализ множественной регрессии

```python
model_df1 = sm.ols(formula="sales_month ~ discount + month + returned", data=df1).fit()
model_df1.summary()
```

```python
model_df2 = sm.ols(formula="sales_month ~ discount", data=df1).fit()
model_df2.summary()
```

```python
model_df3 = sm.ols(formula="sales_month ~ returned", data=df1).fit()
model_df3.summary()
```

```python
fig = px.histogram(df1,
                   x='sales_month',
                   marginal='box',
                   color='returned',
                   color_discrete_sequence=['green','gray'],
                   title='Продажи при условии возврата товара')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
# Сгруппировать данные по годам и месяцам
grouped_data = df1.groupby(['year', 'month']).agg({
    'discount': 'sum',
    'sales_month': 'sum'
}).reset_index()

# Создаем столбец для объединения года и месяца в одну дату
grouped_data['date'] = pd.to_datetime(grouped_data[['year', 'month']].assign(day=1))

# Визуализация
sales_chart = alt.Chart(grouped_data).mark_line(point=True).encode(
    x='date:T',  # Дата будет на оси X
    y='sum(sales_month):Q',  # Продажи будут на оси Y
    tooltip=['date:T', 'sum(sales_month):Q', 'sum(discount):Q'],  # Подсказки
    color=alt.value("steelblue")
).properties(
    title='Динамика продаж и скидок по месяцам',
    width=800,
    height=400
)

# Показать график
sales_chart.display()
```
